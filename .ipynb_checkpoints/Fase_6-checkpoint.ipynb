### Escolhendo a versÃ£o Yolo11s

Essa versÃ£o Ã© mais indicada para treinamento com menos dados

```python
model_path =  "yolo11s.pt"
```

### Caminho do drive com o projeto (Imagens + Labels)

```python
data_yaml = " /content/drive/MyDrive/FIAP_FASE6/data.yaml"
```

#Treinamento 1
### Treinando com 40 Ã©pocas

```python
model = YOLO(model_path)
```

```python
results = model.train(
    data=data_yaml,   # arquivo .yaml com caminhos e classes
    epochs=40,        # nÃºmero de Ã©pocas
    imgsz=640,        # tamanho das imagens
    batch=16,
    name="treino_yolo11",  # nome da pasta de resultados
    device=0
)
```

#Testando o primeiro modelo treinado com imagens de Teste

```python
import os

#Configurando
model_path = "yolo11s.pt"
input_folder = "/content/drive/MyDrive/FIAP_FASE6/test"  # pasta com as imagens
output_folder = "resultados"    # pasta onde salvarÃ¡ as imagens com detecÃ§Ãµes
conf_threshold = 0.25

#InferÃªncia

# Cria a pasta de saÃ­da
os.makedirs(output_folder, exist_ok=True)

model = YOLO(model_path)

# Percorre todos os arquivos da pasta
for filename in os.listdir(input_folder):
    if filename.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp')):
        image_path = os.path.join(input_folder, filename)
        print(f" Processando: {filename}")


        results = model(image_path, conf=conf_threshold)

        output_path = os.path.join(output_folder, f"resultado_{filename}")

        # Salva a imagem com as detecÃ§Ãµes
        results[0].save(filename=output_path)

        print(f"Resultado em: {output_path}")

print("\n InferÃªncia concluÃ­da para todas as imagens!")
```

#Treinamento 2
### Treinando com 60 Ã©pocas

```python
results = model.train(
    data=data_yaml,
    epochs=60,
    imgsz=640,
    batch=16,
    name="treino_yolo11",
    device=0
)
```

#Testando o segundo modelo treinado com imagens de Teste

```python

model_path = "yolo11s.pt"
input_folder = "/content/drive/MyDrive/FIAP_FASE6/test"  # pasta com as imagens
output_folder = "resultados"    # pasta onde salvarÃ¡ as imagens com detecÃ§Ãµes
conf_threshold = 0.25



os.makedirs(output_folder, exist_ok=True)

model = YOLO(model_path)

# Percorre todos os arquivos da pasta
for filename in os.listdir(input_folder):
    if filename.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp')):
        image_path = os.path.join(input_folder, filename)
        print(f" Processando: {filename}")


        results = model(image_path, conf=conf_threshold)

        output_path = os.path.join(output_folder, f"resultado_{filename}")

        results[0].save(filename=output_path)

        print(f"Resultado em: {output_path}")

print("\n InferÃªncia concluÃ­da para todas as imagens!")
```

```python

import matplotlib.pyplot as plt



model_path = "yolo11s.pt"          # modelo (pode ser seu modelo customizado)
input_folder = "/content/drive/MyDrive/FIAP_FASE6/test"     # pasta com imagens
conf_threshold = 0.25              # confianÃ§a mÃ­nima

# Carrega o modelo
model = YOLO(model_path)

# Seleciona duas imagens da pasta
imagens = [f for f in os.listdir(input_folder) if f.lower().endswith(('.jpg', '.jpeg', '.png'))][:2]

# Cria figuras
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

for ax, img_name in zip(axes, imagens):
    img_path = os.path.join(input_folder, img_name)
    results = model(img_path, conf=conf_threshold)

    # Pega a imagem renderizada (com as caixas)
    rendered_img = results[0].plot()  # retorna imagem com bounding boxes desenhadas

    # Mostra no grÃ¡fico
    ax.imshow(rendered_img)
    ax.axis("off")
    ax.set_title(f"DetecÃ§Ãµes: {img_name}")

plt.tight_layout()
plt.show()
```

#AnÃ¡lise

No primeiro treinamento do modelo YOLO11s com 40 Ã©pocas, os resultados mostraram alta precisÃ£o geral de 0.992 e bom desempenho na classe â€œCachorroâ€. O que indicam detecÃ§Ãµes muito consistentes e bem delimitadas. No entanto, o modelo praticamente nÃ£o conseguiu identificar gatos â€” o recall foi 0. Assim, o modelo Ã© muito preciso, mas com baixa cobertura sobre ambas as classes.

JÃ¡ o segundo treinamento, com 60 Ã©pocas apresentou um comportamento mais equilibrado. A precisÃ£o geral caiu levemente para 0.834, mas o recall subiu de 0.5 para 0.75. Isso indica que o modelo passou a detectar mais instÃ¢ncias.
A classe â€œGatoâ€ obteve uma grande melhoria, que passou a ter recall de 0.5./Representando uma evoluÃ§Ã£o em relaÃ§Ã£o ao primeiro treino.

De forma geral o segundo modelo se mostrou superior em equilÃ­brio e cobertura, pois agora identifica bem ambas as classes de forma mais eficiente e demonstrando melhor capacidade de generalizaÃ§Ã£o. Isso demonstra de forma clara como a quantidade de dados e a certa parametrizaÃ§Ã£o dos hiperparametros resulta em modelos melhores.
